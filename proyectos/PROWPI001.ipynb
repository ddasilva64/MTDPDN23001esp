## Paises (extracci√≥n y transformaci√≥n)

```python
# EXTRACCI√ìN DE DATOS
df_pai01 = pd.read_excel('/content/PopulationDataWorldBank.xlsx','Country Metadata')    # Lectura del dataset 01
print('--- Comprobaci√≥n de df_pai01 ---')
print(df_pai01.info)                                                                    # df tiene la calidad necesaria  
df_pai_princ = pd.read_csv('/content/countries_esp.csv')                                # Lectura del dataset exacto en espa√±ol
print('--- Comprobaci√≥n de df_pai_princ ---')
print(df_pai_princ.info)
df_pai_eng = pd.read_csv('/content/countries_eng.csv')                                  # Lectura del dataset exacto en ingl√©s
print('--- Comprobaci√≥n de df_pai_eng ---')
print(df_pai_eng.info)

# TRANSFORMACI√ìN DE DATOS
print('--- Columnas de df_pai01 ---')
for column in df_pai01:                                                                 # Comprobaci√≥n de columnas del df
  print(column)      

df_pai_ext = df_pai01.copy()                                                            # Copia del df para trabajar con √©l
df_pai_ext['Country Code'] = df_pai_ext['Country Code'].str.lower()                     # Estandarizaci√≥n del c√≥digo de poblaci√≥n (iso3) a min√∫sculas

print('--- Columnas de df_pai_eng ---')                                                 # Comprobaci√≥n de columnas del df
for column in df_pai_eng:
  print(column)

print('--- Columnas de df_pai_princ ---')                                               # Comprobaci√≥n de columnas del df
for column in df_pai_princ:
  print(column)

df_pai_princ.drop(['id','alpha2'], axis=1, inplace=True)                                # Limpieza de las columnas no necesarias
                                                                                 
df_pai_princ.set_axis(['codigo_pais',                                                   # Renombrado de columnas para estandarizar nombres
                      'pais'], axis=1, inplace=True)

df_pai01 = pd.merge(df_pai_princ, df_pai_ext, how='inner',                              # Eliminaci√≥n de las filas innecesarias en el df original
                    left_on='codigo_pais', right_on='Country Code')


df_pai01.drop(['Country Code', 'SpecialNotes', 'TableName'],                            # Limpieza de las columnas no necesarias
                    axis=1, inplace=True)  

df_pai01.set_axis(['codigo_pais',                                                       # Renombrado de columnas para estandarizar nombres
                      'pais',
                      'region',
                      'grupo_ingresos'], axis=1, inplace=True)  

print('--- Comprobaci√≥n de df_pai01 ---')
print(df_pai01.info)                                                                    # df filtrado

print('--- Columnas de df_pai01 ---')
for column in df_pai01:                                                                 # Comprobaci√≥n de columnas del df filtrado
  print(column)

'''
  Para poder limpiar y traducir, contenidos de campos, debemos
  agrupar los campos (por contenido), para ver las ocurrencias 
  de cada contenido y cu√°les son esos contenidos
'''
print('--- Regiones a traducir ---')
print(df_pai01.groupby(["region"])["region"].count())                                   # Comprobaci√≥n de regiones (antes de limpiar)

# Substituci√≥n de valores traducidos de las regiones del mundo
df_pai01['region'] = df_pai01['region'].mask(df_pai01['region'] == 'East Asia & Pacific').fillna('Asia oriental y el Pac√≠fico')
df_pai01['region'] = df_pai01['region'].mask(df_pai01['region'] == 'Europe & Central Asia').fillna('Europa y Asia Central')
df_pai01['region'] = df_pai01['region'].mask(df_pai01['region'] == 'Latin America & Caribbean').fillna('Latinoam√©rica y Caribe')
df_pai01['region'] = df_pai01['region'].mask(df_pai01['region'] == 'Middle East & North Africa').fillna('Oriente Medio y Norte de √Åfrica')
df_pai01['region'] = df_pai01['region'].mask(df_pai01['region'] == 'North America').fillna('Am√©rica del Norte')
df_pai01['region'] = df_pai01['region'].mask(df_pai01['region'] == 'South Asia').fillna('Asia del Sur')
df_pai01['region'] = df_pai01['region'].mask(df_pai01['region'] == 'Sub-Saharan Africa').fillna('Africa Sub-sahariana')

print('--- Grupos de ingresos a traducir ---')
print(df_pai01.groupby(["grupo_ingresos"])["grupo_ingresos"].count())                 # Comprobaci√≥n de grupo de ingresos (antes de limpiar)

# Substituci√≥n de valores traducidos de los grupos de ingresos de los paises del mundo
df_pai01['grupo_ingresos'] = df_pai01['grupo_ingresos'].mask(df_pai01['grupo_ingresos'] == 'High income').fillna('Ingresos altos')
df_pai01['grupo_ingresos'] = df_pai01['grupo_ingresos'].mask(df_pai01['grupo_ingresos'] == 'Low income').fillna('Ingresos bajos')
df_pai01['grupo_ingresos'] = df_pai01['grupo_ingresos'].mask(df_pai01['grupo_ingresos'] == 'Lower middle income').fillna('Ingresos medios bajos')
df_pai01['grupo_ingresos'] = df_pai01['grupo_ingresos'].mask(df_pai01['grupo_ingresos'] == 'Upper middle income').fillna('Ingresos medios altos')

print('--- Comprobaci√≥n de df_pai01 ---')
print(df_pai01.info)                                                                # Comprobaci√≥n del df, tras los cambios

print('--- Comprobaci√≥n de NaN en df_pai01 ---')
print(df_pai01.isna().sum())                                                        # Comprobaci√≥n de NaN en el df

print('--- Comprobaci√≥n de nulos en df_pai01 ---')
print(df_pai01.isnull().sum())                                                      # Comprobaci√≥n de nulos en el df
```

```python
# EXTRACCI√ìN DE DATOS
df_sal01_a = pd.read_excel('/content/PopulationDataWorldBank.xlsx','Health1')         # Lectura del dataset 01 (Hoja 1)
print('--- Comprobaci√≥n de df_sal01_a ---')
print(df_sal01_a.info)                                                                # df de Salud - hoja 1
df_sal01_b = pd.read_excel('/content/PopulationDataWorldBank.xlsx','Health2')         # Lectura del dataset 01 (Hoja 2)
print('--- Comprobaci√≥n de df_sal01_b ---')
print(df_sal01_b.info)                                                                # df de Salud - hoja 2
df_sal01 = pd.concat([df_sal01_a,df_sal01_b])   
print('--- Comprobaci√≥n de df_sal01 ---')                                             # df no tiene la calidad necesaria
print(df_sal01.info) 
df_sal02 = pd.read_excel('/content/NHA indicators.xlsx')                              # Lectura del dataset con la calidad requerida
print('--- Comprobaci√≥n final de df_sal02 ---')
print(df_sal02.info)                                                                  # df ya tiene la calidad necesaria

# TRANSFORMACI√ìN DE DATOS

'''
 Queda claro que la OMS no tiene datos disponibles del 2021, pero si del 2020, 
 como habitualmente los datos se recogen en el segundo trimestre, 
 cogeremos los datos del 2020 como si fuesen del 2021.
'''
df_sal02.drop(['2021'], axis=1, inplace=True)                                         # Eliminaci√≥n de las columnas no necesarias

df_sal02.set_axis(['pais', 'indicador', '2021'], axis=1, inplace=True)                # Renombrado de columnas para estandarizar nombres

'''
  Para poder limpiar y traducir, contenidos de campos, debemos
  agrupar los campos (por contenido), para ver las ocurrencias 
  de cada contenido y cu√°les son esos contenidos
'''
print('--- Indicadores a traducir ---')
print(df_sal02.groupby(["indicador"])["indicador"].count())                           # Comprobaci√≥n de indicadores (antes de limpiar)

'''
  Comprobamos que s√≥lo hay dos indicadores que se repitan en la mayor√≠a de paises, 
  que traducidos son:
    - Gasto corriente en salud como %  del PIB
    - Gasto corriente en salud per c√°pita en US$
  M√°s adelante, haremos un arreglo, pra transformas estas dos filas en columnas

  Ahora tenemos el problema de que la columna pa√≠s no nos interesa, 
  en su lugar queremos el c√≥digo del pa√≠s, pero adem√°s,
  el nombre del pa√≠s est√° en ingl√©s. Por lo tanto, hacemos un 
  inner join entre el df con 193 filas y el que tiene m√°s.
  Con esto ya tenemos relacionado, el c√≥digo del pa√≠s iso3 con
  el nombre en ingl√©s del pa√≠s.
'''

df_sal_eng = pd.merge(df_pai_eng, df_sal02, how='inner', left_on='name',                  # Eliminaci√≥n de las filas innecesarias en el df original
                      right_on='pais')
print('--- Paises en ingl√©s con su c√≥digo iso3 ---')
print(df_sal_eng.info)

'''
  Ahora debemos comprobar si, entre los paises que no ha unido, hay alguno que nos interesa.
  Iteramos para comprobar valores incluidos y no incluidos en df_pai_eng.
'''

print('--- Paises contenidos ---')                                                        # Control de filas que ha unido el join
intCont = 0
ntContE = 0
for row in df_sal02.pais:
  strPais = df_sal02.pais.iloc[intCont]
  intCont += 1
  if strPais in df_pai_eng.name.values:
    print(strPais)
    ntContE += 1
print('Total:', ntContE)

print('--- Paises no contenidos ---')                                                     # Control de filas que no ha unido el join
intCont  = 0
ntContNE = 0
for row in df_sal02.pais:
  strPais = df_sal02.pais.iloc[intCont]
  intCont += 1
  if strPais not in df_pai_eng.name.values:
    print(strPais)
    ntContNE += 1
print('Total:', ntContNE)

'''
  Buscamos los paises que no ha podido unir el join y a√±adimos una columna con el nombre 
  ant√≠guo (el que queremos substituir).
  Atenci√≥n:
    Ni Niue, ni las Islas Cook, son estados soberanos reconocidos por la ONU en 2021. 
    Son estados libres asociados a Nueva Zelanda, por lo tanto despreciamos estas filas.
'''

df_filtro01 = df_pai_eng[df_pai_eng['name'].str.contains(r"Cabo V*",                    # B√∫squeda con regex de las filas deseadas    
                            regex=True)] 
df_filtro01.insert(3, "old_name", ['Cabo Verde Republic of'], True)       
df_filtro02 = df_pai_eng[df_pai_eng['name'].str.contains(r"Congo, Democ*" ,     
                            regex=True)]
df_filtro02.insert(3, "old_name", ['Democratic Republic of the Congo'], True)  
df_filtro03 = df_pai_eng[df_pai_eng['name'].str.contains(r"Tanzania*",   
                            regex=True)]
df_filtro03.insert(3, "old_name", ['United Republic of Tanzania'], True)  
df_filtro04 = df_pai_eng[df_pai_eng['name'].str.contains(r"Bolivia*",    
                            regex=True)]
df_filtro04.insert(3, "old_name", ['Bolivia Plurinational States of'], True)  
df_filtro05 = df_pai_eng[df_pai_eng['name'].str.contains("Iran *",       
                            regex=True)]
df_filtro05.insert(3, "old_name", ['Iran'], True)  
df_filtro06 = df_pai_eng[df_pai_eng['name'].str.contains("Czech*",       
                            regex=True)]
df_filtro06.insert(3, "old_name", ['Czech Republic'], True)  
df_filtro07 = df_pai_eng[df_pai_eng['name'].str.contains("Netherlands*", 
                            regex=True)]
df_filtro07.insert(3, "old_name", ['Netherlands (Kingdom of the)'], True) 
df_filtro08 = df_pai_eng[df_pai_eng['name'].str.contains("Moldova*",     
                            regex=True)]
df_filtro08.insert(3, "old_name", ['Republic of Moldova'], True) 
df_filtro09 = df_pai_eng[df_pai_eng['name'].str.contains("North Macedonia*",       
                            regex=True)]
df_filtro09.insert(3, "old_name", ['The Republic of North Macedonia'], True) 
df_filtro10 = df_pai_eng[df_pai_eng['name'].str.contains("United Kingdom*",       
                            regex=True)]
df_filtro10.insert(3, "old_name", ['United Kingdom'], True) 
df_filtro11 = df_pai_eng[df_pai_eng['name'].str.contains("Korea, Republic*",       
                            regex=True)]
df_filtro11.insert(3, "old_name", ['Republic of Korea'], True) 

df_filtro = pd.concat([df_filtro01, df_filtro02, df_filtro03, df_filtro04,                # df con los valores que sustituyen y los sustituidos  
                       df_filtro05, df_filtro06, df_filtro07, df_filtro08, 
                       df_filtro09, df_filtro10, df_filtro11])

print('--- Comprobaci√≥n de que hemos encontrado las filas que no ha podido unir el join ---')
print(df_filtro)

'''
Los paises que transformaremos son:

1. Cabo Verde Republic of => Cabo Verde (2 filas)  
2. Democratic Republic of the Congo  => Congo, Democratic Republic of the (5 filas)  
3. United Republic of Tanzania => Tanzania, United Republic of (2 filas)  
4. Bolivia Plurinational States of  => Bolivia (Plurinational State of) (2 filas)  
5. Iran  => Iran (Islamic Republic of) (2 filas)  
6. Czech Republic  => Czechia (2 filas)  
7. Netherlands (Kingdom of the) => Netherlands (2 filas)
8. Moldova => Moldova, Republic of Republic of (2 filas)
9. North Macedonia => The Republic of North Macedonia (2 filas)
10. United Kingdom => United Kingdom of Great Britain and Northern Ireland  => (2 filas)  
12. Cook Islands (no transformamos)  
13. Niue (no transformamos)  
14. Korea => Korea, Republic of (2 filas)  

Recuento de filas
* Coincidentes 631  
* Total que han hecho match, de las no coincidentes: 25  
* Total que no han hecho match, de las no coincidentes: 5  
* Total: 661 registro originales

Con lo cual cambiamos valores en el df df_sal02, para igualar valores y poder obtener el id correcto
'''

print('--- Substituci√≥n de paises no contenidos ---')                                   # Susbtituci√≥n de los valores no enlazables por los enlazables
intCont=0
for row in df_sal02.pais:
  strPais = df_sal02.pais.iloc[intCont].strip()
  if strPais in df_filtro.old_name.values:
    dfn = df_filtro[df_filtro['old_name'] == strPais]
    nuevo_valor=dfn.name.values.item(0)
    df_sal02.pais.iloc[intCont]=nuevo_valor
  intCont+=1

print('--- Paises no contenidos ---')                                                   # Como paises no enlazados solo tenemos los que hemos rechazado (Islas Cook y Niue)
intCont  = 0
ntContNE = 0
for row in df_sal02.pais:
  strPais = df_sal02.pais.iloc[intCont]
  intCont += 1
  if strPais not in df_pai_eng.name.values:
    print(strPais)
    ntContNE += 1
print('Total:', ntContNE)

'''
  Ahora ya tenemos todos los paises contenidos menos Niue e Islas Cook.
  Por lo tanto podemos limpiar de paises que no interesen en df_sal02, 
  para ello primero haremos un inner join. 
'''

df_sal02 = pd.merge(df_pai_eng, df_sal02, how='inner', left_on='name',                  # Eliminaci√≥n de las filas innecesarias en el df original
                    right_on='pais')

df_sal02.drop(['id', 'alpha2', 'name', 'pais'], axis=1, inplace=True)                   # Eliminaci√≥n de las columnas no necesarias

df_sal02.set_axis(['codigo_pais', 'indicador', '2021'], axis=1, inplace=True)           # Renombrado de columnas para estandarizar nombres 

''' 
  Como ya hemos dicho el indicador que interesa es el
  del Gasto corriente en salud, eliminamos el resto de
  las filas.
'''

df_sal02a=df_sal02.loc[df_sal02['indicador'] == 'Current Health Expenditure (CHE) as % Gross Domestic Product (GDP)']
df_sal02b=df_sal02.loc[df_sal02['indicador'] == 'Current Health Expenditure (CHE) per Capita in US$']

print('--- Indicadores a traducir df_sala ---')
print(df_sal02a.groupby(["indicador"])["indicador"].count())                             # Comprobaci√≥n de indicadores (antes de limpiar)

print('--- Indicadores a traducir df_salb ---')
print(df_sal02b.groupby(["indicador"])["indicador"].count())                             # Comprobaci√≥n de indicadores (antes de limpiar)

'''
  Ahora tenemos 185 paises de los 193 que consideramos 
  con 2 indicadores v√°lidos para el an√°lisis posterior
'''
df_sal02a.set_axis(['codigo_pais', 'indicador',                                         # Renombrado de columnas para estandarizar nombres 
                      'gasto_corriente_salud_%_pib'], 
                   axis=1, inplace=True) 

df_sal02a.drop(['indicador'], axis=1, inplace=True)                                     # Eliminaci√≥n de las columnas no necesarias

df_sal02b.set_axis(['codigo_pais', 'indicador',                                         # Renombrado de columnas para estandarizar nombres 
                      'gasto_corriente_salud_pc'], 
                   axis=1, inplace=True) 

df_sal02b.drop(['indicador'], axis=1, inplace=True)                                     # Eliminaci√≥n de las columnas no necesarias

df_sal02 = pd.merge(df_sal02a, df_sal02b, how='inner',                                  # Eliminaci√≥n de las filas innecesarias en el df original
                    left_on='codigo_pais', right_on='codigo_pais')

print('--- Comprobaci√≥n final de df_sal02 ---')
print(df_sal02.info)                                                                    # df ya tiene la calidad necesaria y est√° l√≠mpio
```
